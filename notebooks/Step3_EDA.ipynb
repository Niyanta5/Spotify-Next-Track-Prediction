```python
import lyricsgenius
import time

# Initialize Genius API with an increased timeout to prevent timeouts
genius = lyricsgenius.Genius(token, timeout=30)

# Loop through DataFrames from df31 to df40
for i in range(31, 41):
    df_name = f"df{i}"
    df = globals().get(df_name)
    
    if df is not None:
        print(f"Processing {df_name}...")
        
        # Create a new column to store the lyrics if it doesn't exist
        if 'lyrics' not in df.columns:
            df['lyrics'] = None  # Initialize the column with None

        # Loop through each row in the DataFrame to fetch lyrics
        for index, row in df.iterrows():
            song_title = row['track_track_name']
            song_artist = row['track_artist_name']
            print(f"Fetching lyrics for: {song_title} by {song_artist}")
            
            retries = 3  # Number of retries
            for attempt in range(retries):
                try:
                    # Search for the song using the track name and artist name
                    song_info = genius.search_song(title=song_title, artist=song_artist)

                    # Fetch and store lyrics if found
                    if song_info:
                        df.at[index, 'lyrics'] = song_info.lyrics  # Store lyrics in the new column
                        print(f"Lyrics for '{song_info.title}' stored.\n")
                    else:
                        print(f"Lyrics not found for {song_title} by {song_artist}\n")
                    break  # Break the retry loop if successful
                    
                except Exception as e:
                    print(f"Error fetching lyrics: {e}")
                    if attempt < retries - 1:
                        wait_time = 2 ** (attempt + 1)  # Exponential backoff: 2, 4, 8 seconds
                        print(f"Retrying in {wait_time} seconds...")
                        time.sleep(wait_time)  # Wait before retrying
                    else:
                        print("Max retries reached. Moving to the next song.\n")
                        df.at[index, 'lyrics'] = None  # Optionally set to None if failed

            # Sleep between requests to avoid rate limits (e.g., 3 seconds)
            time.sleep(3)

        # Update the DataFrame in globals
        globals()[df_name] = df
        print(f"Completed processing {df_name}\n")

# Display the first few rows of df31 as an example
df31.head()
```

```python
import pandas as pd
import string
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download stopwords if not already available
#nltk.download('stopwords')
#nltk.download('punkt')

# Load English stop words
stop_words = set(stopwords.words('english'))

# Define a function to preprocess the lyrics and create a bag of words
def create_bag_of_words(lyrics):
    if pd.isna(lyrics):  # Handle missing lyrics
        return None

    # Convert to lowercase
    lyrics = lyrics.lower()

    # Remove punctuation
    lyrics = lyrics.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the text into words
    words = word_tokenize(lyrics)

    # Remove stop words
    filtered_words = [word for word in words if word not in stop_words and word.isalpha()]

    # Count the frequency of each word
    word_counts = Counter(filtered_words)

    # Get the 10 most common words
    most_common_words = [word for word, count in word_counts.most_common(10)]

    # Return as a space-separated string (bag of words)
    return ' '.join(most_common_words)


df['bag_of_words'] = df['lyrics'].apply(create_bag_of_words)

# Display the resulting DataFrame with the new 'bag_of_words' column
df[['track_track_name', 'lyrics', 'bag_of_words']].head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.sentiment import SentimentIntensityAnalyzer

# Sample DataFrame setup (Assuming df already exists with 'lyrics' and 'bag_of_words' columns)
# Example: df = pd.DataFrame({'lyrics': [...], 'bag_of_words': [...]})

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis on the 'bag_of_words' column
df['sentiment_bag_of_words'] = df['bag_of_words'].fillna('').apply(lambda words: sia.polarity_scores(words)['compound'])

# Initialize TF-IDF Vectorizer for bag_of_words
tfidf_vectorizer = TfidfVectorizer(max_features=500, stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['bag_of_words'].fillna(''))

# Convert to DataFrame for analysis
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Combine with original DataFrame
df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)

# Display the updated DataFrame with sentiment and TF-IDF features
df[['track_track_name', 'bag_of_words', 'sentiment_bag_of_words']].head(10)
```

```python
# Creating a sentiment category based on sentiment score
def categorize_sentiment(score):
    if score > 0.05:
        return 'positive'
    elif score < -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment_category'] = df['sentiment_bag_of_words'].apply(categorize_sentiment)
```

```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load your dataset
df = pd.read_csv("your_dataset.csv")

# Initialize BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_bert_embeddings(text):
    # Tokenize input
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    # Generate embeddings
    with torch.no_grad():
        outputs = model(**inputs)
    # Get the embeddings of [CLS] token
    return outputs.last_hidden_state[:, 0, :].numpy()

# Generate embeddings for each song in the dataset
df['lyrics_embeddings'] = df['lyrics'].apply(lambda x: get_bert_embeddings(x) if pd.notnull(x) else np.zeros((1, 768)))

# Define recommendation function
def recommend_next_song(current_song_idx, df, top_n=5):
    current_embedding = df.iloc[current_song_idx]['lyrics_embeddings']
    embeddings = np.vstack(df['lyrics_embeddings'].values)
    
    # Calculate cosine similarity
    similarities = cosine_similarity(current_embedding.reshape(1, -1), embeddings).flatten()
    
    # Get the indices of the top_n similar songs, excluding the current song
    similar_indices = similarities.argsort()[::-1][1:top_n+1]
    
    return df.iloc[similar_indices][['track_track_name', 'track_artist_name', 'lyrics']]

# Example usage: recommend next song based on the song at index 0
recommended_songs = recommend_next_song(0, df)
print(recommended_songs)
```
